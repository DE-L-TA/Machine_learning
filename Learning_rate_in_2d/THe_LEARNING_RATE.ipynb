```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def gradient_descent(derivative_func,initial_guess,multiplier=0.0001,prescion=0.0001,max_iter=100):
    new_x=initial_guess
    x_list=[new_x]
    defrencial_list=[derivative_func(new_x)]
    
    for i in range(max_iter):
        previous_x=new_x
        gradient=derivative_func(previous_x)
        new_x=previous_x - multiplier*gradient
        if(i==10):
            print(df(previous_x))
        x_list.append(new_x)
        defrencial_list.append(gradient)
        step_size = abs(new_x - previous_x)
        
        if(step_size < prescion):
            break;
            
    return new_x , x_list ,defrencial_list
```

```python
def df(x):
    return 4*x**3 - 8*x
```

```python
n=50
low_gamma = gradient_descent(derivative_func=df,initial_guess=1.9,multiplier=0.0001 ,max_iter=n)
mid_gamma = gradient_descent(derivative_func=df,initial_guess=1.9,multiplier=0.0002 ,max_iter=n)
high_gamma = gradient_descent(derivative_func=df,initial_guess=1.9,multiplier=0.001 ,max_iter=n)
iteration_list=list(range(0,n+1))
```

```python
plt.figure(figsize=[20, 10])
plt.xlim(0,n)
plt.ylim(0,50)
low_values=np.array(low_gamma[1])                    
                    
plt.plot(iteration_list,df(np.array(mid_gamma[1])),color='cyan')
plt.scatter(iteration_list,df(np.array(mid_gamma[1])),s=80,color='cyan')
plt.annotate('Low Learning Rate',(10,8.762967353240745),color='cyan',fontsize=30,xytext=(30,43))



plt.plot(iteration_list,df(low_values))
plt.scatter(iteration_list,df(low_values),s=80)
plt.annotate('High Learning Rate',(10,8.762967353240745),color='blue',fontsize=30,xytext=(30,46))
plt.plot(iteration_list,df(np.array(high_gamma[1])),color='red')
plt.scatter(iteration_list,df(np.array(high_gamma[1])),s=80,color='red')
plt.annotate('Good Learning Rate',(10,8.762967353240745),color='red',fontsize=30,xytext=(30,40))
plt.show()  
```
