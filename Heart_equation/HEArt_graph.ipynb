## HEART EQUATION
  $$ x = 16 sin(t)^3 $$
  $$ y = -(13math.cos(t)-5math.cos(2t)-2*math.cos(3t)-math.cos(4t)) $$

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
def x(t):
    return -16*math.sin(t)**3  
```

```python
def y(t):
    return (13*math.cos(t)-5*math.cos(2*t)-2*math.cos(3*t)-math.cos(4*t))
```

```python
plt.figure(figsize=[10,10])
i=0
while(i<=math.pi):
    if(i>0 and i<1.5):
        plt.scatter(x(i),y(i),color='red',s=150)
    else:
        plt.scatter(x(i),y(i),color='red',s=150)
    i=i+0.01
j=0    
while(j>=-math.pi):
    if(j<0 and j>-1.5):
        plt.scatter(x(j),y(j),color='red',s=50)
    else:
        plt.scatter(x(j),y(j),color='red',s=50)
    j=j-0.01    
plt.xlim(-17,17)    
plt.show()
```

```python
plt.figure(figsize=[10,10])
i=0
while(i<= 2*math.pi ):
    if((i>0 and i<1.6) or (i<2*math.pi and i>1.5+math.pi)):
        plt.scatter(x(i),y(i),color='orange',s=150)
    elif((i>2.2 and i<math.pi) or (i>math.pi and i<0.9+math.pi)):
        plt.scatter(x(i),y(i),color='green',s=50)
    elif((i>1.65 and i<2.18) or (i>.92+math.pi and i<1.45+math.pi)):
        plt.scatter(x(i),y(i),color='purple',s=100)
    else:
        plt.scatter(x(i),y(i),color='white',s=100)
    i=i+0.01

plt.xlim(-17,17)    
plt.show()
```
